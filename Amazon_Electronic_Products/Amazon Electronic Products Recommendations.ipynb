```python
# Convert all entries to strings first
df['no_of_ratings'] = df['no_of_ratings'].astype(str)
df['discount_price'] = df['discount_price'].astype(str)
df['actual_price'] = df['actual_price'].astype(str)

# Identify non-numeric values in the problematic columns
print(df['no_of_ratings'][~df['no_of_ratings'].str.replace(',', '').str.isnumeric()])
print(df['discount_price'][~df['discount_price'].str.replace('₹', '').str.replace(',', '').str.isnumeric()])
print(df['actual_price'][~df['actual_price'].str.replace('₹', '').str.replace(',', '').str.isnumeric()])
```

## Clean Data

```python
# Remove rows with non-numeric values in 'no_of_ratings'
df = df[df['no_of_ratings'].str.replace(',', '').str.isnumeric()]

# Remove rows with non-numeric values in 'discount_price' and 'actual_price'
df = df[df['discount_price'].str.replace('₹', '').str.replace(',', '').str.isnumeric()]
df = df[df['actual_price'].str.replace('₹', '').str.replace(',', '').str.isnumeric()]

# Now convert the cleaned columns to numeric
df['no_of_ratings'] = df['no_of_ratings'].str.replace(',', '').astype(float)
df['discount_price'] = df['discount_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Show the first few rows to confirm the changes
df.head()
```

```python
# Sort products by number of ratings (popularity)
top_products = df.sort_values('no_of_ratings', ascending=False)

# Show the top 5 products
top_5_products = top_products.head(5)
print(top_5_products[['name', 'discount_price', 'no_of_ratings']])
```

```python
def recommend_top_products(df, num_products=5):
    # Sort products by the number of ratings
    sorted_products = df.sort_values('no_of_ratings', ascending=False)
    
    # Get the top products
    top_products = sorted_products.head(num_products)
    
    return top_products[['name', 'discount_price', 'no_of_ratings']]

# Get the top 5 recommended products
recommendations = recommend_top_products(df)
print("Top 5 Product Recommendations:")
print(recommendations)
```

```python
# Get random products as a comparison
random_products = df.sample(n=5)
print("Random Product Recommendations:")
print(random_products[['name', 'discount_price', 'ratings']])
```

```python
# Calculate the revenue for each product (assuming each rating is a purchase)
df['revenue'] = df['discount_price'] * df['no_of_ratings']

# Sort products by revenue to identify top-selling products
top_selling_products = df.sort_values('revenue', ascending=False).head(5)
print("Top 5 Top-Selling Products:")
print(top_selling_products[['name', 'discount_price', 'revenue']])

# Compare with top recommended products
recommended_products = recommend_top_products(df, num_products=5)

# Calculate precision: how many of the recommended products are also top-selling
matched_products = pd.merge(recommended_products, top_selling_products, on='name')
precision = len(matched_products) / len(recommended_products)
print(f"Precision of the recommendation system: {precision:.2f}")

# Output the recommended and top-selling products for comparison
print("Recommended Products vs. Top-Selling Products:")
print(pd.concat([recommended_products[['name']], top_selling_products[['name']]], axis=1))
```

## Exploratory Data Analysis (EDA) on Customer Purchase History

## 1.1. Distribution of Prices

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of discount prices
plt.figure(figsize=(10, 6))
sns.histplot(df['discount_price'], bins=30, kde=True)
plt.title('Distribution of Discount Prices')
plt.xlabel('Discount Price (₹)')
plt.ylabel('Frequency')
plt.show()
```

## 1.2. Distribution of Number of Ratings

```python
# Plot the distribution of number of ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['no_of_ratings'], bins=30, kde=True, color='orange')
plt.title('Distribution of Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Frequency')
plt.show()
```

## EDA on Top-Selling Products

## 1.3 Top 10 Products by Revenue

```python
# Plot the top 10 products by revenue
top_10_revenue_products = df.sort_values('revenue', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='revenue', y='name', data=top_10_revenue_products, palette='viridis')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue (₹)')
plt.ylabel('Product Name')
plt.show()
```

## 1.4 Comparison of Revenue with Number of Ratings

```python
# Plot revenue vs number of ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x='no_of_ratings', y='revenue', data=df, hue='discount_price', palette='coolwarm', size='discount_price')
plt.title('Revenue vs Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Revenue (₹)')
plt.legend(title='Discount Price', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
